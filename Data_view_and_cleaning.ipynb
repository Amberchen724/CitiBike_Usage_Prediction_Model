Clean LEHDR data

```python
LEHDR = pd.read_csv('data/LEHD-resident.csv')
LEHDR.head()
```

```python
LEHDR['male'] = LEHDR.male/LEHDR.total
LEHDR['age_under_29'] = LEHDR.age_under_29/LEHDR.total
LEHDR['age_30_54'] = LEHDR.age_30_54/LEHDR.total
LEHDR['age_55_over'] = LEHDR.age_55_over/LEHDR.total
LEHDR['white'] = LEHDR.white/LEHDR.total
LEHDR['black'] = LEHDR.black/LEHDR.total
LEHDR['asian'] = LEHDR.asian/LEHDR.total
LEHDR['others'] = LEHDR.others/LEHDR.total
LEHDR['two_or_more'] = LEHDR.two_or_more/LEHDR.total
LEHDR['less_high_school'] = LEHDR.less_high_school/LEHDR.total
LEHDR['high_school'] = LEHDR.high_school/LEHDR.total
LEHDR['college'] = LEHDR.college/LEHDR.total
LEHDR['bachelor'] = LEHDR.bachelor/LEHDR.total
```

```python
LEHDR.drop(['female'], axis=1, inplace=True)
```

```python
LEHDR.head()
```

```python
LEHDR.to_csv('Clean_Data/LEHDR.csv', index=False)
```
